```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("players_data-2024_2025.csv")
df.fillna(0, inplace=True)
```

```python
top_tackles = df[['Player', 'Tkl']].sort_values(by='Tkl', ascending=False).head(5).reset_index(drop=True)
top_tackles_won = df[['Player', 'TklW']].sort_values(by='TklW', ascending=False).head(5).reset_index(drop=True)
top_tackle_stats = (
    df[["Player", "Tkl", "TklW"]]
    .sort_values(by="Tkl", ascending=False)
    .head(5)
    .reset_index(drop=True))
top_tackle_stats["Win %"] = (top_tackle_stats["TklW"] / top_tackle_stats["Tkl"]) * 100
top_tackle_stats["Win %"] = top_tackle_stats["Win %"].round(1)
top_tackle_stats
```

```python
x = np.arange(len(top_tackle_stats))
width = 0.25

plt.figure(figsize=(12, 6))

plt.bar(x - width, top_tackle_stats["Tkl"], width, label='Tackles', color='lightblue', edgecolor="black")
plt.bar(x, top_tackle_stats["TklW"], width, label='Tackles Won', color='orange', edgecolor="black")
plt.bar(x + width, top_tackle_stats["Win %"], width, label='Win %', color='lightgreen', edgecolor="black")

plt.xticks(x, top_tackle_stats["Player"])
plt.ylabel("Values")
plt.title("Top 5 Players – Tackles, Tackle Wins & Win Percentage")
plt.legend()
plt.tight_layout()
plt.show()
```
