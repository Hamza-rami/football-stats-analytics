## Clean Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("Data.csv", sep=';', encoding='latin1')
df.fillna(0, inplace=True)
```

```python
top_scorers = df.sort_values(by="Goals", ascending=False).head(3)
top_scorers[['Player', 'Goals']]
plt.bar(top_scorers["Player"], top_scorers["Goals"], color="lightgreen", edgecolor="black")
```
