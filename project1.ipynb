## Clean Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("players_data-2024_2025.csv")
# df.fillna(0, inplace=True)
```

```python
Comp = df['Comp'].value_counts().head(5).index
df_top5 = df[df['Comp'].isin(Comp)]
age_stats_top5 = df_top5.groupby('Comp')['Age'].agg(['min','max','mean'])
age_stats_top5.rename(columns={'min':'Min Age','max':'Max Age','mean':'Average Age'}, inplace=True)
age_stats_top5
```

## Min, Average, and Max Age by Top 5 Comp

```python
Comp = age_stats_top5.index
min_age = age_stats_top5['Min Age']
max_age = age_stats_top5['Max Age']
avg_age = age_stats_top5['Average Age']

x = np.arange(len(Comp))
width = 0.25

plt.figure(figsize=(10,6))
plt.bar(x - width, min_age, width, label='Min Age', color='skyblue', edgecolor="black")
plt.bar(x, avg_age, width, label='Average Age', color='orange', edgecolor="black")
plt.bar(x + width, max_age, width, label='Max Age', color='lightgreen', edgecolor="black")

# Labels
plt.xlabel('Comp')
plt.ylabel('Age')
plt.title('Min, Average, and Max Age by Top 5 Comp')
plt.xticks(x, Comp)
plt.legend()
plt.show()
```
