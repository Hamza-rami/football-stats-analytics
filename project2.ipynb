## Top 5 Players by Goals, Assists, and Combined Contribution

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# import data
df = pd.read_csv("players_data-2024_2025.csv")
df.fillna(0, inplace=True)
```

```python
top_goal_scorers = df[['Player', 'Gls']].sort_values(by='Gls', ascending=False).head(5).reset_index(drop=True)
top_assists = df[['Player', 'Ast']].sort_values(by='Ast', ascending=False).head(5).reset_index(drop=True)
top_goal_and_assists = df[['Player', 'G+A']].sort_values(by='G+A', ascending=False).head(5).reset_index(drop=True)
top_goal_and_assists
```

```python
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Top 5 Goal Scorers
ax[0].bar(top_goal_scorers["Player"], top_goal_scorers["Gls"], color="#87CEEB", edgecolor="black")
ax[0].set_xticks(range(len(top_goal_scorers)))
ax[0].set_xticklabels(top_goal_scorers["Player"], rotation=45, ha='right')
ax[0].set_title("Top 5 Goal Scorers")

# Top 5 Assist Providers
ax[1].bar(top_assists["Player"], top_assists["Ast"], color="#87CEEB", edgecolor="black")
ax[1].set_xticks(range(len(top_assists)))
ax[1].set_xticklabels(top_assists["Player"], rotation=45, ha='right')
ax[1].set_title("Top 5 Assist Providers")

# Top 5 Goals + Assists
ax[2].bar(top_goal_and_assists["Player"], top_goal_and_assists['G+A'], color="#87CEEB", edgecolor="black")
ax[2].set_xticks(range(len(top_goal_and_assists)))
ax[2].set_xticklabels(top_goal_and_assists["Player"], rotation=45, ha='right')
ax[2].set_title("Top 5 Goals + Assists")

plt.tight_layout()
plt.subplots_adjust(wspace=0.3)
plt.show()

```
